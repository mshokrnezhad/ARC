```python
import arckit
import cairosvg
import os
from IPython.display import Image, display

folder = 'figs/train/'
```

```python
train_set, _ = arckit.load_data()
```

```python
for i, task in enumerate(train_set[:10]):
    print(i, task)
    
print(train_set[0])
```

```python
task = arckit.load_single('00d62c1b') # You can also load tasks by ID
print(len(task.train), len(task.test))

task.train[0][0] # input

# task.train[0][1] # output
```

```python
for i, task in enumerate(train_set):
    drawing = arckit.draw_task(task, width=20, height=12, include_test=True)
    arckit.output_drawing(drawing, f"{folder}train_{i}.svg")
```

```python
for i in range(0, 401):
    svg_filepath = os.path.join(folder, f'train_{i}.svg')
    png_filepath = os.path.join(folder, f'train_{i}.png')
    if os.path.exists(svg_filepath):
        cairosvg.svg2png(url=svg_filepath, write_to=png_filepath)
    else:
        print(f"File {svg_filepath} does not exist.")
```

```python
for i in range(0, 10):
    png_filepath = os.path.join(folder, f'train_{i}.png')
    if os.path.exists(png_filepath):
        display(Image(filename=png_filepath))
    else:
        print(f"File {png_filepath} does not exist.")
```
